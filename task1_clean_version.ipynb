# Task 1 - Clean Version Using Utility Functions

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from dstapi import DstApi

# Import utility functions
from house_price_utils import (
    download_house_prices, clean_house_data, filter_complete_series,
    reindex_to_base, download_cpi, calculate_real_prices,
    plot_time_series, plot_ranking
)
```

## Question 1.a - Nominal House Prices

```python
# Download, clean, filter and reindex in 4 lines!
df_raw = download_house_prices()
df = clean_house_data(df_raw)
df = filter_complete_series(df)
df = reindex_to_base(df)
```

```python
# Plot time series
plot_time_series(df)
```

```python
# Plot ranking
plot_ranking(df)
```

## Question 1.b - Real House Prices

```python
# Download CPI and calculate real prices in 2 lines!
df_cpi = download_cpi()
df_real = calculate_real_prices(df, df_cpi)
```

```python
# Plot real prices
plot_time_series(df_real, index_col='Real_Index_1992', title_suffix=' - Realpriser')
```

```python
# Plot real price ranking
plot_ranking(df_real, index_col='Real_Index_1992')
```

## Comparison: Nominal vs Real Prices

Compare the growth in one line:

```python
# Compare nominal vs real for København Ejerlejligheder
comparison = df_real[
    (df_real['Landsdel'] == 'Landsdel Byen København') & 
    (df_real['Ejendomskategori'] == 'Ejerlejligheder, i alt')
][['Kvartal', 'Indeks_1992', 'Real_Index_1992']].tail(1)

print(f"København Ejerlejligheder - Latest Quarter ({comparison['Kvartal'].values[0]}):")
print(f"  Nominal growth: {comparison['Indeks_1992'].values[0]:.0f}%")
print(f"  Real growth: {comparison['Real_Index_1992'].values[0]:.0f}%")
print(f"  Inflation effect: {comparison['Indeks_1992'].values[0] - comparison['Real_Index_1992'].values[0]:.0f} percentage points")
```
